```python
from tools import *
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import pickle

phrase="J'ai adoré cette application magnifique"

def prediction_user(phrase):
    #Load vocabulary from TFIDF 
    with open("data/feature.pkl", "rb") as f:
        voca = pickle.load(f)
    transformer = TfidfVectorizer(vocabulary = voca )
    # Featuring de la phrase user
    user = transformer.fit_transform([nett(phrase)])
    # Load classifier
    cls=pickle.load(open("data/clf.pkl", "rb"))
    # Predict from loaded classifier
    #cls.predict(user),cls.predict_proba(user).max()
    return cls.predict(user)

prediction_user(phrase)
```
